```python
from pathlib import Path
import os

# Change this if your Drive folder name/path is different
PROJECT_ROOT = Path("/content/drive/MyDrive/retrofit")

mydrive = Path("/content/drive/MyDrive")
if not mydrive.exists():
    raise RuntimeError("Google Drive is not mounted. Run Cell 2 first (drive.mount).")

if not PROJECT_ROOT.exists():
    print("PROJECT_ROOT not found:", PROJECT_ROOT)
    print("Available folders in MyDrive (top 50):")
    for p in sorted(mydrive.iterdir())[:50]:
        if p.is_dir():
            print(" -", p.name)
    raise FileNotFoundError("Set PROJECT_ROOT to the correct folder inside /content/drive/MyDrive")

os.chdir(PROJECT_ROOT)
print("CWD:", Path.cwd())

DATA_PATH = str(PROJECT_ROOT / "data/raw/bridge_data.csv")
NROWS = 5000

SEVERITY_COL = "Severity"
RETROFIT_COL = "Suggested_Retrofit"
```

```python
import pandas as pd
from pathlib import Path

data_file = Path(DATA_PATH)
if not data_file.exists():
    raise FileNotFoundError(f"CSV not found at: {data_file}")

# Load a sample first (fast, safe for large datasets)
df = pd.read_csv(data_file, nrows=NROWS)
print('Sample shape:', df.shape)
display(df.head())
```

```python
print(df.dtypes)

missing_pct = (df.isna().mean().sort_values(ascending=False) * 100).round(2)
print('Missing % (top 25):')
display(missing_pct.head(25))
```

```python
for col in [SEVERITY_COL, RETROFIT_COL]:
    if col not in df.columns:
        print(f'Column not found: {col}')
        continue
    print(f'Value counts for {col}:')
    display(df[col].astype(str).value_counts().head(25))
```
